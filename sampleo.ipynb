```python
import json
import pandas as pd
```

# Carga de datos

```python
with open('data.txt', 'r') as data:
    subgraph_raw = data.read()

subgraph_raw = subgraph_raw.split('\n')

subgraph = []
for s in subgraph_raw:
    subgraph.append(s.split('\t'))

subgraph
```

```python
data_raw = pd.DataFrame(subgraph)
data_raw.drop_duplicates(inplace=True)
ent_1 = data_raw[0].to_list()
rel = data_raw[1].to_list()
ent_2 = data_raw[2].to_list()
```

```python
subgraph = []
for i in range(0,len(ent_1)):
    subgraph.append([ent_1[i], rel[i], ent_2[i]])
subgraph
```

```python
with open('traduciones_entidades.json', 'r') as traducciones_json:
    traducciones = json.load(traducciones_json)
```

Depuro las lineas que no tienen traduccion y traduzco el resto.

```python

subgraph_fb = []
subgraph_null = []

for relation in subgraph:
    
    try:
        entidad_1 = relation[0]
        entidad_2 = relation[2]
        relacin  = relation[1]

        entidad_1_t = traducciones[entidad_1]
        entidad_2_t = traducciones[entidad_2]

        subgraph_fb.append([entidad_1_t,relacin,entidad_2_t])

    except:
        subgraph_null.append(relation)
```

```python
with open('relations.dict', 'r') as data:
    relaciones_raw = data.read()

relaciones_raw = relaciones_raw.split('\n')

relaciones = []

for r in relaciones_raw:
    try:
        r = r.split('\t')
        relaciones.append(r[1].split("/")[1:])

    except:
        print(r)

relaciones
```

Elimino las lineas repetidas.

```python
subgraph_fb_depurado = []
for subgraph in subgraph_fb:
    if subgraph[1].split("/")[0]!='_':
        subgraph_fb_depurado.append(subgraph)
```

Genero un sample con todas las profesiones.

```python
nacionalidades = []
for subgraph in subgraph_fb_depurado:
    if subgraph[1].split("/")[1]=='people' and subgraph[1].split("/")[-1]=='nationality':
        nacionalidades.append(subgraph[2])

nacionalidades = list(set(nacionalidades))

nacionalidades
```

Genero subset por profesion.

```python
nacionalidades_dict = {}

for nacionalidad in nacionalidades:
    
    lista_profesionales = []
    
    for subgraph in subgraph_fb_depurado:

        if subgraph[1].split("/")[-1]=='nationality' and subgraph[2] == nacionalidad:
            lista_profesionales.append(subgraph)
    
    if len(lista_profesionales) > 1:
        nacionalidades_dict[nacionalidad] = lista_profesionales

nacionalidades_dict

```

```python
uk_ls = nacionalidades_dict["United Kingdom"][:100]
us_ls =  nacionalidades_dict["United States"][:100]
```
